```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
# Load data from JSON files

with open('adagrad_results.json', 'r') as f:
    adagrad = json.load(f)

with open('adam_results.json', 'r') as f:
    adam = json.load(f)

with open('lasso_adagrad_results.json', 'r') as f:
    lasso_adagrad = json.load(f)

with open('lasso_adam_results.json', 'r') as f:
    lasso_adam = json.load(f)

with open('lasso_momentum_results.json', 'r') as f:
    lasso_momentum = json.load(f)

with open('lasso_plain_gd_results.json', 'r') as f:
    lasso_plain_gd = json.load(f)

with open('lasso_rmsprop_results.json', 'r') as f:
    lasso_rmsprop = json.load(f)

with open('lasso_sklearn_results.json', 'r') as f:
    lasso_sklearn = json.load(f)

with open('momentum_results.json', 'r') as f:
    momentum = json.load(f)

with open('rmsprop_results.json', 'r') as f:
    rmsprop = json.load(f)

with open('sgd_adagrad_results.json', 'r') as f:
    sgd_adagrad = json.load(f)

with open('sgd_adam_results.json', 'r') as f:
    sgd_adam = json.load(f)

with open('sgd_momentum_results.json', 'r') as f:
    sgd_momentum = json.load(f)

with open('sgd_rmsprop_results.json', 'r') as f:
    sgd_rmsprop = json.load(f)

with open('sgd_results.json', 'r') as f:
    sgd = json.load(f)


# Separating results
ridge_adagrad_mse = adagrad['MSE Ridge']
ols_adagrad_mse = adagrad['MSE OLS']
ridge_adagrad_r2 = adagrad['R2 Ridge']
ols_adagrad_r2 = adagrad['R2 OLS']

ridge_adam_mse = adam['MSE Ridge']
ols_adam_mse = adam['MSE OLS']
ridge_adam_r2 = adam['R2 Ridge']
ols_adam_r2 = adam['R2 OLS']

lasso_adagrad_mse = lasso_adagrad['MSE GD Lasso']
lasso_adagrad_r2 = lasso_adagrad['R2 GD Lasso']

lasso_adam_mse = lasso_adam['MSE GD Lasso']
lasso_adam_r2 = lasso_adam['R2 GD Lasso']

lasso_momentum_mse = lasso_momentum['MSE GD Lasso']
lasso_momentum_r2 = lasso_momentum['R2 GD Lasso']

lasso_plain_gd_mse = lasso_plain_gd['MSE GD Lasso']
lasso_plain_gd_r2 = lasso_plain_gd['R2 GD Lasso']

lasso_rmsprop_mse = lasso_rmsprop['MSE GD Lasso']
lasso_rmsprop_r2 = lasso_rmsprop['R2 GD Lasso']

lasso_sklearn_mse = lasso_sklearn['MSE']
lasso_sklearn_r2 = lasso_sklearn['R2']

ridge_momentum_mse = momentum['MSE Ridge']
ridge_momentum_r2 = momentum['R2 Ridge']
ols_momentum_mse = momentum['MSE OLS']
ols_momentum_r2 = momentum['R2 OLS']

ridge_rmsprop_mse = rmsprop['MSE Ridge']
ridge_rmsprop_r2 = rmsprop['R2 Ridge']
ols_rmsprop_mse = rmsprop['MSE OLS']
ols_rmsprop_r2 = rmsprop['R2 OLS']

ols_adagrad_sgd_mse = sgd_adagrad['MSE GD OLS']
ridge_adagrad_sgd_mse = sgd_adagrad['MSE GD Ridge']
lasso_adagrad_sgd_mse = sgd_adagrad['MSE GD Lasso']
ols_adagrad_sgd_r2 = sgd_adagrad['R2 GD OLS']
ridge_adagrad_sgd_r2 = sgd_adagrad['R2 GD Ridge']
lasso_adagrad_sgd_r2 = sgd_adagrad['R2 GD Lasso']

ols_adam_sgd_mse = sgd_adam['MSE GD OLS']
ridge_adam_sgd_mse = sgd_adam['MSE GD Ridge']
lasso_adam_sgd_mse = sgd_adam['MSE GD Lasso']
ols_adam_sgd_r2 = sgd_adam['R2 GD OLS']
ridge_adam_sgd_r2 = sgd_adam['R2 GD Ridge']
lasso_adam_sgd_r2 = sgd_adam['R2 GD Lasso']

ols_momentum_sgd_mse = sgd_momentum['MSE GD OLS']
ridge_momentum_sgd_mse = sgd_momentum['MSE GD Ridge']
lasso_momentum_sgd_mse = sgd_momentum['MSE GD Lasso']
ols_momentum_sgd_r2 = sgd_momentum['R2 GD OLS']
ridge_momentum_sgd_r2 = sgd_momentum['R2 GD Ridge']
lasso_momentum_sgd_r2 = sgd_momentum['R2 GD Lasso']

ols_sgd_mse = sgd['MSE GD OLS']
ridge_sgd_mse = sgd['MSE GD Ridge']
lasso_sgd_mse = sgd['MSE GD Lasso']
ols_sgd_r2 = sgd['R2 GD OLS']
ridge_sgd_r2 = sgd['R2 GD Ridge']
lasso_sgd_r2 = sgd['R2 GD Lasso']

ols_rmsprop_sgd_mse = sgd_rmsprop['MSE GD OLS']
ridge_rmsprop_sgd_mse = sgd_rmsprop['MSE GD Ridge']
lasso_rmsprop_sgd_mse = sgd_rmsprop['MSE GD Lasso']
ols_rmsprop_sgd_r2 = sgd_rmsprop['R2 GD OLS']
ridge_rmsprop_sgd_r2 = sgd_rmsprop['R2 GD Ridge']
lasso_rmsprop_sgd_r2 = sgd_rmsprop['R2 GD Lasso']
```

```python
# Plotting results

# Prepare data for heatmap
optimizers = ['ADAGRAD', 'ADAM', 'MOMENTUM', 'RMSPROP']
models = ['OLS', 'RIDGE', 'LASSO']
mse_data = [
    [ols_adagrad_mse, ridge_adagrad_mse, lasso_adagrad_mse],
    [ols_adam_mse, ridge_adam_mse, lasso_adam_mse],
    [ols_momentum_mse, ridge_momentum_mse, lasso_momentum_mse],
    [ols_rmsprop_mse, ridge_rmsprop_mse, lasso_rmsprop_mse]
]
mse_array = np.array(mse_data)

fig, ax = plt.subplots(figsize=(6, 4))
im = ax.imshow(mse_array, cmap='viridis')

# Show all ticks and label them
ax.set_xticks(np.arange(len(models)))
ax.set_yticks(np.arange(len(optimizers)))
ax.set_xticklabels(models)
ax.set_yticklabels(optimizers)

# Rotate the tick labels and set alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(optimizers)):
    for j in range(len(models)):
        ax.text(j, i, f"{mse_array[i, j]:.4f}", ha="center", va="center", color="w")

fig.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()

# Prepare data for heatmap
optimizers = ['ADAGRAD', 'ADAM', 'MOMENTUM', 'RMSPROP']
models = ['OLS', 'RIDGE', 'LASSO']
mse_data = [
    [ols_adagrad_r2, ridge_adagrad_r2, lasso_adagrad_r2],
    [ols_adam_r2, ridge_adam_r2, lasso_adam_r2],
    [ols_momentum_r2, ridge_momentum_r2, lasso_momentum_r2],
    [ols_rmsprop_r2, ridge_rmsprop_r2, lasso_rmsprop_r2]
]
mse_array = np.array(mse_data)

fig, ax = plt.subplots(figsize=(6, 4))
im = ax.imshow(mse_array, cmap='viridis')

# Show all ticks and label them
ax.set_xticks(np.arange(len(models)))
ax.set_yticks(np.arange(len(optimizers)))
ax.set_xticklabels(models)
ax.set_yticklabels(optimizers)

# Rotate the tick labels and set alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(optimizers)):
    for j in range(len(models)):
        ax.text(j, i, f"{mse_array[i, j]:.4f}", ha="center", va="center", color="w")

fig.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()
```
